```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io

% matplotlib inline
```

Read all files in the dataset

```python
keys = ['pause', 'play', 'volup', 'voldown']

def load_data():
    data = {
        'pause': [],
        'play': [],
        'volup': [],
        'voldown': [],
        'none': []
    }
    
    for key in data:
        col = io.imread_collection('dataset/' + key + '*')
        col = [cv2.cvtColor(im, cv2.COLOR_RGB2GRAY) for im in col]
        data[key] = list(col)
    
    return data
    
data = load_data()
```

```python
df = pd.DataFrame(data)
df.info()
```


## Preprocess images

### applying HOG

```python
from skimage.feature import hog
from skimage import exposure

def apply_hog(img, visualise=True):
    fd, hog_im = hog(img, orientations=9, pixels_per_cell=(5, 5),
                    cells_per_block=(1, 1), visualise=visualise, block_norm='L2-Hys')
    
    hog_image_rescaled = exposure.rescale_intensity(hog_im, in_range=(0, 10))
    return hog_image_rescaled
    
    
# test function
sample = df['volup'][1]

hog_im = apply_hog(sample)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(sample, cmap=plt.cm.gray)
ax1.set_title('Input image')
ax1.set_adjustable('box-forced')

ax2.axis('off')
ax2.imshow(hog_im, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
ax1.set_adjustable('box-forced')
plt.show()
```

```python
# try to reduce noise by adding circular mask
def add_disk_mask(im):
    r, c = im.shape
    rw, cl = np.ogrid[:r, :c]
    cr, cc = r / 2, c / 2
    
    mask = (((rw - cr) ** 2) + ((cl - cc) ** 2) > (cr ** 2))
    im[mask] = 0
    
    return im

fin = add_disk_mask(hog_im)
io.imshow(fin)
```

#### vectorise images

```python
def flatten(img):
    return img.reshape(img.shape[0] * img.shape[1])
```

```python
im = flatten(fin)
print(im.shape)
im1 = im[im == 0]
print(im1.shape)
sparsity = im1.shape[0] / im.shape[0] * 100
sparsity
```

sparsity looks goood. so, make em a sparse matrix

```python
from scipy.sparse import csr_matrix

def sparsify(vec):
    # take a row vector and make sparse matrix
    return csr_matrix(vec)

sp = sparsify(im)
```

```python
def preprocessing_pipeline(image):
    print('processing...')
    final = apply_hog(image)
    final = add_disk_mask(final)
    final = flatten(final)
#     final = sparsify(final)
    
    return final
```

this is gonna take a while ...

```python
from multiprocessing import Pool

num_partitions = 10 #number of partitions to split dataframe
num_cores = 4 #number of cores on your machine

def parallelize_dataframe(df, func):
    df_split = np.array_split(df, num_partitions)
    pool = Pool(num_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df

def apply_wrapper(data):
    return data.applymap(preprocessing_pipeline)
```

```python
# processed_df = df.applymap(preprocessing_pipeline)
processed_df = parallelize_dataframe(df, apply_wrapper)
processed_df.info()
```

```python
print(type(processed_df['pause'][0]))
processed_df.head()
```

looks like it worked

#### convert dataframes to arrrays for sklearn

```python
def df_to_nparray(df):
    m = len(df)
    targets = np.array(df.keys())
    
    X, Y = [], []
    
    for label in targets:
        temp = df[label].values
        l = len(temp)
        X.extend(temp)
        Y.extend([label] * l)

    X = np.array(X)
    Y = np.array(Y)
    
    return X, Y

X, Y = df_to_nparray(processed_df)
targets = list(df.keys())
```

## Building a good predictive model for the dataset

use labelizers on the targets

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(targets)
print(le.classes_)
Y = le.transform(Y)
```

get training and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
from sklearn import svm

svm_clf = svm.SVC(gamma=0.001)
svm_clf.fit(X_test, Y_test)
print('accuracy : ', svm_clf.score(X_test, Y_test) * 100, '%')
svm_clf
```

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, targets):
    print(cm)
    
    plt.imshow(cm, interpolation=None, cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(targets))
    plt.xticks(tick_marks, targets)
    plt.yticks(tick_marks, targets)
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

Y_pred = svm_clf.predict(X_test)
plot_confusion_matrix(confusion_matrix(Y_test, Y_pred), targets)
```
